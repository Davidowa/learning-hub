```python
# ================================
# EJEMPLOS DE IF, IF ELSE Y IF/ELIF/ELSE EN PYTHON
# ================================

# ---------
# 1. IF
# ---------
# El código dentro del if solo se ejecuta
# si la condición es verdadera (True)

edad = 20

if edad >= 18:
    print("IF: Eres mayor de edad")

# Si la condición fuera falsa, no se imprimiría nada


# ---------
# 2. IF ELSE
# ---------
# Se usa cuando queremos dos caminos:
# uno si la condición es verdadera y otro si es falsa

edad = 15

if edad >= 18:
    print("IF ELSE: Puedes entrar")
else:
    print("IF ELSE: No puedes entrar")


# ---------
# 3. IF / ELIF / ELSE
# ---------
# Se usa cuando hay varias condiciones posibles
# Python evalúa de arriba hacia abajo

nota = 7

if nota >= 9:
    print("IF ELIF ELSE: Excelente")
elif nota >= 6:
    print("IF ELIF ELSE: Aprobado")
else:
    print("IF ELIF ELSE: Reprobado")


# ---------
# EJEMPLO EXTRA CON STRINGS
# ---------

color = "rojo"

if color == "verde":
    print("El color es verde")
elif color == "rojo":
    print("El color es rojo")
else:
    print("Color no reconocido")
```

```python
# ================================
# EJEMPLO DE MATCH / CASE EN PYTHON
# ================================
# match se usa como alternativa a if / elif / else
# cuando queremos comparar un valor con varios casos posibles

# ---------
# EJEMPLO 1: MATCH CON NÚMEROS
# ---------

opcion = 2

match opcion:
    case 1:
        print("Elegiste la opción 1")
    case 2:
        print("Elegiste la opción 2")
    case 3:
        print("Elegiste la opción 3")
    case _:
        # El guion bajo (_) funciona como "else"
        print("Opción no válida")


# ---------
# EJEMPLO 2: MATCH CON STRINGS
# ---------

color = "azul"

match color:
    case "rojo":
        print("El color es rojo")
    case "verde":
        print("El color es verde")
    case "azul":
        print("El color es azul")
    case _:
        print("Color no reconocido")


# ---------
# EJEMPLO 3: MATCH CON VARIAS OPCIONES EN UN CASO
# ---------

dia = "sabado"

match dia:
    case "lunes" | "martes" | "miercoles" | "jueves" | "viernes":
        print("Es un día laboral")
    case "sabado" | "domingo":
        print("Es fin de semana")
    case _:
        print("Día no válido")
```

```python
# ================================
# EJEMPLOS DE FOR Y WHILE EN PYTHON
# ================================

# ---------
# 1. BUCLE FOR CON range(inicio, fin, paso)
# ---------
# range(0, 10, 1) significa:
# - empezar en 0
# - terminar en 10 (sin incluirlo)
# - avanzar de 1 en 1

for i in range(0, 10, 1):
    print("FOR:", i)


# ---------
# 2. BUCLE FOR CON OTRO PASO
# ---------
# En este caso avanza de 2 en 2

for i in range(0, 11, 2):
    print("FOR con paso 2:", i)


# ---------
# 3. BUCLE WHILE
# ---------
# while se ejecuta mientras la condición sea verdadera
# Hay que tener cuidado de modificar la variable
# para evitar bucles infinitos

contador = 0

while contador < 5:
    print("WHILE:", contador)
    contador += 1  # aumentamos el contador en 1


# ---------
# 4. EJEMPLO COMPARATIVO FOR VS WHILE
# ---------

# FOR: cuando sabemos cuántas veces se repetirá
for i in range(0, 5, 1):
    print("FOR comparativo:", i)

# WHILE: cuando no sabemos exactamente cuántas veces
contador = 0
while contador < 5:
    print("WHILE comparativo:", contador)
    contador += 1
```

```python
# Ejemplo de codigo documentado porque use IA
# Programa que muestra los números narcisistas del 1 al 1000
# Un número narcisista es aquel que es igual a la suma de sus dígitos
# elevados a la potencia del número de dígitos que tiene

# Código documentado con ayuda de ChatGPT (versión GPT-5.2)

print("Los numeros narcicistas del 1 al 1000 son:")

# Bucle que recorre los números del 1 al 1000
for num in range(1, 1001, 1):

    # Convertimos el número a texto para poder recorrer sus dígitos
    numero_texto = str(num)

    # Calculamos cuántos dígitos tiene el número
    longitud = len(numero_texto)

    # Variable donde se guardará la suma de los dígitos elevados a la potencia
    suma = 0

    # Bucle para recorrer cada dígito del número
    for digito in range(0, longitud, 1):

        # Convertimos cada dígito a entero y lo elevamos
        # a la potencia del total de dígitos
        suma += int(numero_texto[digito]) ** longitud

    # Si la suma es igual al número original,
    # entonces es un número narcisista
    if suma == num:
        print(num, end=" ")

# Salto de línea para mejorar la salida
print(end="\n")

# Mensaje final
print("Esos serian los numeros :)")
```

```python
# Realizar un programa que solicite la cantidad de alumnos a registrar.
# Luego, para cada alumno, ingresar su promedio.
# El programa debe mostrar la lista de alumnos junto con sus respectivos promedios.
# Finalmente, debe calcular y mostrar el promedio general del grupo.

cantidad_registros = int(input("Ingresa la cantidad de alumnos a registrar: "))
alumnos = []
promedios = []

suma_promedios_op1 = 0
suma_promedios_op2 = 0

for i in range(1, cantidad_registros + 1, 1):
  # Acercamiento 1 - Append directo del input
  # alumnos.append(input(f"Ingresa el nombre del alumno {i}: "))
  # promedios.append(float(input(f"Ingresa el promedio del alumno {i}: ")))

  # Acercamiento 2 - Uso de variables auxiliares
  alumno = input(f"Ingresa el nombre del alumno {i}: ")
  promedio = float(input(f"Ingresa el promedio del alumno {i}: "))

  suma_promedios_op1 += promedio

  alumnos.append(alumno)
  promedios.append(promedio)

promedio_op1 = suma_promedios_op1/cantidad_registros

print("\nLista de alumnos y promedios:")
for i in range(0, len(alumnos), 1):
  print(f"{alumnos[i]}    \t {promedios[i]}")
  suma_promedios_op2 += promedios[i]

promedio_op2 = suma_promedios_op2/cantidad_registros

print(f"El promedio general del grupo es: {promedio_op1}")
print(f"El promedio general del grupo es: {promedio_op2}")

# Poner el promedio con dos decimales
print(f"El promedio general del grupo es: {promedio_op2:.2f}")
```

```python
# Ejercicio 01 - 09 Feb 2026

# 1) Control de acceso
# Tu programa debe pedir:
# nombre (texto)
# edad (número entero)
# credencial (True/False)
# Crea una función llamada (por ejemplo) puede_entrar(edad, credencial) que devuelva True o False.
# Regla para entrar:
# Entra solo si edad >= 18 y credencial == True
# Guarda cada intento en una lista.
# Cada intento debe guardar mínimo: nombre, edad, credencial y resultado (si entra o no).
# Salida esperada:

# Mostrar si la persona entra o no.
# Al final, mostrar la lista de intentos registrados.

# Definición de funcion
def Puede_Entrar (edad, credencial):
  if edad >= 18 and credencial == True:
    return True
  else:
    return False

# Acercamiento de listas
nombre = []
edad = []
credencial = []
puede_entrar = []

# Este si se usa en ambos
intentos_lista = []
intentos_diccionario = []

# Preguntar cuantos intentos quiero hacer
cant_intentos = int(input("Cuantos intentos quieres hacer: "))

for i in range(1, cant_intentos + 1, 1):
  print(f"Intento {i}")
  ingresa_nombre = input("Ingresa tu nombre: ")
  ingresa_edad = int(input("Ingresa tu edad: "))
  ingresa_credencial = input("¿Tienes credencial? (Si/No): ")

  # lower() es una funcion de python de los strings para hacerlos minusculas
  if ingresa_credencial.lower() == "si":
    ingresa_credencial = True
  else:
    ingresa_credencial = False

  # Acercamiento Listas
  nombre.append(ingresa_nombre)
  edad.append(ingresa_edad)
  credencial.append(ingresa_credencial)
  puede_entrar.append(Puede_Entrar(ingresa_edad, ingresa_credencial))
  #puede_entrar.append(Puede_Entrar(credencial[i-1], credencial[i-1])) # Si quiero obtener el valor de las listas

  # Acercamiento de Diccionario
  datos = {
      "nombre": "",
      "edad": 0,
      "credencial": False,
      "puede_entrar": False
  }
  datos["nombre"] = ingresa_nombre
  datos["edad"] = ingresa_edad
  datos["credencial"] = ingresa_credencial
  datos["puede_entrar"] = Puede_Entrar(ingresa_edad, ingresa_credencial)

  # Imprimir que si entro o no
  if puede_entrar[i - 1] == True:
    print("Acercamiento Listas: Si entro")
  else:
    print("Acercamiento Listas: No entro")

  if datos["puede_entrar"] == True:
    print("Acercamiento Diccionario: Si entro")
  else:
    print("Acercamiento Diccionario: No entro")

  # Almacenar Historico Acercamiento listas
  intentos_lista.append((ingresa_nombre, ingresa_edad, ingresa_credencial, Puede_Entrar(ingresa_edad, ingresa_credencial)))
  intentos_diccionario.append(datos)

# Impresion de Historico
for i in range(0, cant_intentos, 1):
  print(f"{nombre[i]} - {edad[i]} - {credencial[i]} - {puede_entrar[i]}")
print(intentos_lista)
print(intentos_diccionario)
```
