```python
# Quiero un programa
# Que el estudiante agregue las califacionaciones de su semestre
# Me saque el promedio de calificacion
# Me indique cuales reprobo
# Me indique cuantas paso
# Y un porcentaje
# USANDO LISTAS

calificaciones = [] # Una lista vacia, es decir no tiene elementos

programa_corriendo = True # bool

while programa_corriendo == True:
  calificacion = float(input("Ingrese calificación de materia: "))
  calificaciones.append(calificacion)
  continuar = input("Desea agregar otra calificación? (s/n) ")
  if continuar == "n":
    programa_corriendo = False # break - break the saca del ciclo
print("Tus calificaciones ingresadas son: ")
print(calificaciones)

# Para saber el tamaño de una lista se usa len - lenght - tamaño
cantidad_calificaciones = len(calificaciones)
cantidad_reprobadas = 0
cantidad_aprobadas = 0
suma_calificaciones = 0
# Ciclo de acceso
for i in range(0, cantidad_calificaciones, 1):
  if calificaciones[i] >= 6.0:
    cantidad_aprobadas = cantidad_aprobadas + 1
  else:
    cantidad_reprobadas = cantidad_reprobadas + 1
  suma_calificaciones = suma_calificaciones + calificaciones[i]

print(f"Promedio: {suma_calificaciones/cantidad_calificaciones}")
print(f"Aprobadas: {cantidad_aprobadas}")
print(f"Reprobadas: {cantidad_reprobadas}")
```
