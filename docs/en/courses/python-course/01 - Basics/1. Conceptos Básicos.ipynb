```python
# Ejercicio
# Realiza el calculo del area de un rectagulo y el calculo del area de un
# triangulo y decir cual area es mayor.

# Solucion
base_rectangulo = float(input("Ingrese la base del rectangulo: "))
altura_rectangulo = float(input("Ingrese la altura del rectangulo: "))
base_triangulo = float(input("Ingrese la base del triangulo: "))
altura_triangulo = float(input("Ingrese la altura del triangulo: "))

area_rectangulo = (base_rectangulo * altura_rectangulo)
area_triangulo = (base_triangulo * altura_triangulo) / 2

print(f"El area del rectangulo es {area_rectangulo}")
print(f"El area del rectangulo es {area_triangulo}")

if area_rectangulo > area_triangulo:
  print("El area del rectangulo es mayor")
else:
  print("El area del triangulo es mayor")
```

```python
# If Elif Else -> Si esto se cumple haz esto, si no haz esto otro, si no haz esto
# Cond es una condición
#            |
#          /   \
#    Si - /Cond1\ - No
#     |   \     /   |
#     |     \ /     |
#  AcciónSi1         |
#                   |
#                  /   \
#            Si - /Cond2\ - No
#             |   \     /   |
#             |     \ /     |
#           AcciónSi2      ...
#                           |
#                         /   \
#                   Si - /CondN\ - No
#                    |   \     /   |
#                    |     \ /     |
#                 AcciónSiN       AcciónNo <- Caso base

# Si tu calificación es mayor o igual que 9.0 entonces es A
# Si tu calificación es mayor o igual que 8.0 entonces es B
# Si tu calificación es mayor o igual que 7.0 entonces es C
# Si tu calificación es mayor o igual que 6.0 entonces es D
# Si tu calificación es menor que 6.0 es F

calificacion_alumno = 10

if calificacion_alumno >= 9.0:
  print("A")
elif calificacion_alumno >= 8.0:
  print("B")
elif calificacion_alumno >= 7.0:
  print("C")
elif calificacion_alumno >= 6.0:
  print("D")
else:
  print("F") # Caso base, es decir no cumpli la condiciones de arriba, entonces
             # es esta.
```

```python
# Retos - 7 Aprobación de Contratación
# Solicita al usuario:

#     Años de experiencia.
#     Resultado de entrevista (0–100).
#     Expectativa salarial.

# Condiciones:

#     (Aprobado si experiencia es mayor o igual a 3) y
#     (entrevista es mayor o igual a 85) y
#     (expectativa salarial es menor o igual a 45000)

#     En revisión si
#     (entrevista es mayor o igual a 80) y
#     [
#        (experiencia es mayor o igual a 2) o
#        (expectativa salarial es menor o igual a 45000).
#     ]
#     Rechazado en cualquier otro caso.

# Muestra la decisión.

anios_exp = float(input("Ingrese sus años de experiencia: "))
resultado_entrevista = int(input("Ingrese el resultado de la entrevista (0-100)"))
expectativa_salarial = float(input("Ingrese la expectativa salarial: "))

if anios_exp >= 3 and resultado_entrevista >= 85 and expectativa_salarial <= 45000:
  print("Aprobado")
elif resultado_entrevista >= 80 and (anios_exp >= 2 or expectativa_salarial <= 45000):
  print("En revisión")
else:
  print("Rechazado")
```
