```python
# ============================================================
# BLOQUE 6) if/else: mayoría de edad (>= 18)
# ============================================================
print("Explicación:")
print("- Se usa un comparador: >=")
print("- Si Edad >= 18: 'puedo votar'. Si no: 'no puedo votar'.")
print("\nCódigo (como texto):")
print("Edad = 19")
print("if (Edad >= 18):")
print('    print("puedo votar")')
print("else:")
print('    print("no puedo votar")')
print("\nResultado:")

Edad = 19

if (Edad >= 18):
    print("puedo votar")
else:
    print("no puedo votar")
```

```python
# ============================================================
# BLOQUE 7) elif: seleccionar mensaje según compañía
# ============================================================
print("Explicación:")
print("- elif se usa cuando hay múltiples caminos posibles.")
print("- Se compara un texto (compomparador) contra opciones.")
print("\nCódigo (como texto):")
print('ATT = "WOW!"')
print('MOVISTAR = "SIGUE EXISTIENDO?"')
print('TELCEL = "HAY SEÑAL?"')
print('compomparador = "ATT"')
print("if (compomparador == 'ATT'):")
print("    print(ATT)")
print("elif (compomparador == 'MOVISTAR'):")
print("    print(MOVISTAR)")
print("elif (compomparador == 'TELCEL'):")
print("    print(TELCEL)")
print("else:")
print('    print("no hay datos")')
print('print("bien hecho")')
print("\nResultado:")

ATT = "WOW!"
MOVISTAR = "SIGUE EXISTIENDO?"
TELCEL = "HAY SEÑAL?"

compomparador = "ATT"

if (compomparador == "ATT"):
    print(ATT)
elif (compomparador == "MOVISTAR"):
    print(MOVISTAR)
elif (compomparador == "TELCEL"):
    print(TELCEL)
else:
    print("no hay datos")

print("bien hecho")
```

```python
# ============================================================
# BLOQUE 8) Input + elif: mensajes por edad
# <18 menor de edad, <21 menor de edad en USA, >=21 mayor mundialmente
# ============================================================
print("Explicación:")
print("- input() regresa texto, por eso se usa int(...) para convertir a número.")
print("- Reglas:")
print("  - <18: menor de edad")
print("  - <21: menor de edad en USA")
print("  - >=21: mayor de edad mundialmente")
print("\nCódigo (como texto):")
print('edad = int(input("ingresa tu edad: "))')
print("if (edad < 18):")
print('    print("eres menor de edad")')
print("elif (edad < 21):")
print('    print("eres menor de edad en usa")')
print("else:")
print('    print("eres mayor de edad mundialmente")')
print("\nResultado:")

edad = int(input("ingresa tu edad: "))
if (edad < 18):
    print("eres menor de edad")
elif (edad < 21):
    print("eres menor de edad en usa")
else:
    print("eres mayor de edad mundialmente")
```

```python
# ============================================================
# BLOQUE 9) Input + rangos con and: semestre
# 1-3 bienvenido, 4-6 mitad, 7-8 último, otro incorrecto
# ============================================================
print("Explicación:")
print("- Se revisa un rango usando 'and'.")
print("- Reglas:")
print("  - 1 a 3: bienvenido a la universidad")
print("  - 4 a 6: mitad del camino")
print("  - 7 a 8: último jalón")
print("  - otro: valor incorrecto")
print("\nCódigo (como texto):")
print('semestre = int(input("cual es tu semestre: "))')
print("if (semestre >= 1 and semestre <= 3):")
print('    print("buenvenido a la universidad")')
print("elif (semestre <= 6):")
print('    print("ya vas a mitad del camino ")')
print("elif (semestre <= 8):")
print('    print("ya es el ultimo jalon")')
print("else:")
print('    print("no me dijiste un valor correcto")')
print("\nResultado:")

semestre = int(input("cual es tu semestre: "))

if (semestre >= 1 and semestre <= 3):
    print("buenvenido a la universidad")
elif (semestre <= 6):
    print("ya vas a mitad del camino ")
elif (semestre <= 8):
    print("ya es el ultimo jalon")
else:
    print("no me dijiste un valor correcto")
```
